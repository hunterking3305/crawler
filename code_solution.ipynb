```python
import requests
import shutil
rs = requests.session()
res = rs.get("https://fbfh.trade.gov.tw/rich/text/common/code_98/CheckImageCode.aspx",stream=True, verify=False)
f = open("chick.png","wb")
shutil.copyfileobj(res.raw, f)
f.close()
```
